### DHIS2 metadata extraction



```python
response = session.get(f"{URL}/api/dataElements", params={"paging" : False})
```

```python
objects = response.json()["dataElements"]
```

```python
data_elements = pd.DataFrame.from_records(objects)
data_elements.head()
```

```python
response = session.get(
    f"{URL}/api/dataSets", 
    params={"paging" : False , "fields": "id,name,periodType,createdBy,dataSetElements"}, 
    )
```

```python
objects = response.json()["dataSets"]
datasets = pd.DataFrame.from_records(objects)
datasets.head()
```

```python
objects_cleaned = []
for obj in objects:

    dataset_elements = []
    for element in obj.get("dataSetElements"):
        dataset_elements.append(element["dataElement"]["id"])

    
    new_object={
        "name":obj.get("name"),
        "id":obj.get("id"),
        "lastUpdated":obj.get("lastUpdated"),
        "periodType":obj.get("periodType"),
        "dataSetElements": dataset_elements
    }

    objects_cleaned.append(new_object)

```

```python
objects_cleaned[0]
datasets = pd.DataFrame.from_records(objects_cleaned)
datasets.head()
```

```python
response = session.get(f"{URL}/api/schemas/dataSet")
for prop in response.json()["properties"]:
    print(
        prop.get("name"), prop.get("fieldName")
    )
```
